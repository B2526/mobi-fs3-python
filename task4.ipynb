**<a id="#">Task 4.1.2.</a>** Implement horizontal and vertical 3&times;3 Prewitt filters:

```python
def prewitt_h(img_in):
    h, w = img_in.shape
    img_out = numpy.zeros_like(img_in,dtype=float)
    kernel = numpy.array([
        [-1, 0, 1],
        [-1, 0, 1],
        [-1, 0, 1]
    ])

    for y in range(1, h - 1): 
        for x in range(1, w - 1):
            neighborhood = img_in[y-1:y+2, x-1:x+2]
            img_out[y,x] = numpy.sum(neigborhood * kernel)
    return img_out


def prewitt_v(img_in):
    h, w = img_in.shape
    img_out = numpy.zeros_like(img_in,dtype=float)
    kernel = numpy.array([
        [-1, -1, -1],
        [0, 0, 0],
        [1, 1, 1]
    ])

    for y in range(1, h - 1): 
        for x in range(1, w - 1):
            neighborhood = img_in[y-1:y+2, x-1:x+2]
            img_out[y,x] = numpy.sum(neigborhood * kernel)
    return img_out
```
